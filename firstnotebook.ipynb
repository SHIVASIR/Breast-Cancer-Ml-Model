### SPLIT THE DATA SET

```python
df.sample(3)
```

```python
x = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

```python
x.shape
```

```python
y.shape
```

```python
# spliting the dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2 , random_state = 0)
```

```python
x_train.shape
```

```python
x_test.shape
```

### FEATURING SCALING 

```python
from sklearn.preprocessing import StandardScaler
```

```python
Scaler = StandardScaler()
```

```python
x_train = Scaler.fit_transform(x_train)
x_test = Scaler.transform(x_test)
```

### Building the model

## 1) Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(random_state = 0)
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
lr.coef_## slope of the best fit line
```

```python
lr.intercept_# intercept of the slpoe
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix , f1_score , precision_score , recall_score
```

```python
acc= accuracy_score(y_test , y_pred)
f1 = f1_score(y_test , y_pred)
prec = precision_score(y_test , y_pred)
rec = recall_score(y_test , y_pred)
```

```python
result = pd.DataFrame([['Logistic Regression ',acc,f1,prec,rec]],
                      columns = ['Model','Accuracy','F1 Score','Precision','Recall'])
```

```python
result
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

### Cross Validation

```python
from sklearn.model_selection import cross_val_score
```

```python
accuracy = cross_val_score(estimator = lr,X=x_train,y=y_train,cv=10)
```

```python
Accuracy1= format(accuracy.mean()*100)
print(f"Accuracy score of Linear Regression  is : {Accuracy1}")
print("Standard Deviation is : {:.2f} %".format(accuracy.std()*100))
```

### 2) Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier_rm =RandomForestClassifier(random_state =0)
classifier_rm.fit(x_train ,y_train)
```

```python
y_pred = classifier_rm.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix , f1_score , precision_score , recall_score
```

```python
acc= accuracy_score(y_test , y_pred)
f1 = f1_score(y_test , y_pred)
prec = precision_score(y_test , y_pred)
rec = recall_score(y_test , y_pred)
```

```python
model_result = pd.DataFrame([['Random Forest ',acc,f1,prec,rec]],
                      columns = ['Model','Accuracy','F1 Score','Precision','Recall'])
```

```python
result = pd.concat([result, model_result], ignore_index=True)
```

```python
result
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

### Cross Validation for Random Forest

```python
from sklearn.model_selection import cross_val_score
```

```python
accuracy = cross_val_score(estimator = classifier_rm,X=x_train,y=y_train,cv=10)
Accuracy2 = format(accuracy.mean()*100)
print(f"Accuracy score of random forest is : {Accuracy2}")
print("Standard Deviation is : {:.2f} %".format(accuracy.std()*100))
```

```python
# Checking the best accuracy to select the model
if Accuracy2 < Accuracy1:
    best_model = Accuracy1
    print("Using Linear Regression")
else:
    best_model = Accuracy2
    print("Using Random Forest")
```
